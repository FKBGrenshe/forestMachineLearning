```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.metrics import accuracy_score
```

```python
# load dataset
data = load_iris()
x = data.data
y = data.target
```

```python
x.shape
```

```python
x[1]
```

```python
# 随机森林
rf = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=3, oob_score=True)
rf.fit(x, y)
```

```python
# 孤立森林
iso_forest = IsolationForest(n_estimators=100, contamination=0.1, random_state=42)
iso_forest.fit(x)
```

```python
y_pred = rf.predict(x)
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"OOB Score: {rf.oob_score_ * 100:.2f}%")
```

```python
# visualization
import matplotlib.pyplot as plt
plt.bar(range(x.shape[1]), rf.feature_importances_)
plt.xticks(range(x.shape[1]), data.feature_names, rotation=45)
plt.ylabel('Feature Importance')
plt.title('Random Forest Feature Importance (Iris)')
plt.tight_layout()
```
